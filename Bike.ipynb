```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('day.csv')
data.head()
```

```python
temp = data.copy()
```

```python
# loop to change the season column numerical var into categorical val
for i in range(len(temp['season'])):
    if temp['season'][i]==1:
        temp['season'][i]='Spring'
    elif temp['season'][i]==2:
        temp['season'][i]='Summer'
    elif temp['season'][i]==3:
        temp['season'][i]='Fall'
    elif temp['season'][i]==4:
        temp['season'][i]='Winter'
    else:
        temp['season'][i]='nan'
        
# loop to change the month column numerical var into categorical val 
    
for i in range(len(temp['mnth'])):
    if temp['mnth'][i]==1:
        temp['mnth'][i]='Jan'
    elif temp['mnth'][i]==2:
        temp['mnth'][i]='Feb'
    elif temp['mnth'][i]==3:
        temp['mnth'][i]='Mar'
    elif temp['mnth'][i]==4:
        temp['mnth'][i]='Apr'
    elif temp['mnth'][i]==5:
        temp['mnth'][i]='May'
    elif temp['mnth'][i]==6:
        temp['mnth'][i]='Jun'
    elif temp['mnth'][i]==7:
        temp['mnth'][i]='Jul'
    elif temp['mnth'][i]==8:
        temp['mnth'][i]='Aug'
    elif temp['mnth'][i]==9:
        temp['mnth'][i]='Sep'
    elif temp['mnth'][i]==10:
        temp['mnth'][i]='Oct'
    elif temp['mnth'][i]==11:
        temp['mnth'][i]='Nov'
    elif temp['mnth'][i]==12:
        temp['mnth'][i]='Dec'
    else:
        temp['mnth'][i]='NaN'
```

```python
temp.head()
```

```python
#applying one hot encoder
xt=temp[['season','mnth','workingday','atemp','hum','windspeed']]
y = temp.cnt
X = pd.get_dummies(xt)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(X_train,y_train)
```

```python
print("Making prediction on the following staring 5 ")
#print(X_test.head())
print("Predictions are")
print(model.predict(X_test.head()))
```

```python
from sklearn.metrics import mean_absolute_error
%matplotlib inline
print(mean_absolute_error(y_test.head(),model.predict(X_test.head())))
```

```python
def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, targ_train)
    preds_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, preds_val)
    return(mae)

for i in [5,50,500,5000]:
    my_mae = get_mae(i,X_train,X_test,y_train,y_test)
    print("Max Leaf node %d \t MAE: %d" % (i,my_mae))
```

```python
from sklearn.ensemble import RandomForestRegressor

rfreg = RandomForestRegressor()

rfreg.fit(X_train,y_train)

rfmae = mean_absolute_error(y_test,rfreg.predict(X_test))
print("RF MAE: ",rfmae)
```
